```python
navn = ['Tidspunkt_Faktisk_Ankomst_Holdeplass_Fra','Tidspunkt_Faktisk_Avgang_Holdeplass_Fra',
       'Tidspunkt_Planlagt_Ankomst_Holdeplass_Fra', 'Tidspunkt_Planlagt_Avgang_Holdeplass_Fra']
timeStamp = df[[navn[0],navn[1],navn[2],navn[3]]]
```

```python
for i in range(len(navn)):
    for j in range(timeStamp.shape[0]):
        timeStamp[navn[i]][j] = timeStamp[navn[i]][j].split(':')
        for tall in range(len(timeStamp[navn[i]][j])):
            if (timeStamp[navn[i]][j][tall] == ''):
                timeStamp[navn[i]][j][tall] = 0
            else : 
                timeStamp[navn[i]][j][tall] = int(timeStamp[navn[i]][j][tall])
```

```python
def timeCheck(timeTable):
    changeIndex = []
    for rad in range(len(timeTable)):
        if(type(timeTable[rad][0]) == str):
            print(f"place {rad} is still a string")
        if (timeTable[rad][0] > 23):
            diff = timeTable[rad][0]-24
            timeTable[rad][0] = 0+diff
            changeIndex.append(rad)
    return changeIndex
```
